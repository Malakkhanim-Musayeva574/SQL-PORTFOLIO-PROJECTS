```python
# E-Commerce Sales Analysis SQL
## In this portfolio project , l analyzed a cleaned e-commerce dataset containing retail transactions.
## The goal is to extract actionable business insights regarding product performance,regional sales,and profitability using advanced SQL techniques.
```

```python
## Which product categories generate the highest sales volume and profit?


select 
    product_category,
    SUM(quantity_sold) as total_units_sold,
    ROUND(SUM(total_revenue),2) as total_sales_ravenue,
    ROUND(SUM(profit),2) as total_net_profit
FROM commerce
GROUP BY product_category
ORDER BY total_sales_revenue DESC;
```

```python
## What are the preferred payment methods across different global regions?

select
    customer_region,
    payment_method,
    COUNT(order_id) as total_transactions,
    ROUND(SUM(total_revenue),2) as regional_ravenue
FROM commerce
GROUP BY customer_region, payment_method
ORDER BY customer_region, regional_revenue DESC;
```

```python
## What are our top-tier single orders based on profitability?

WITH RankedOrders AS (
    SELECT
         order_id,
         product_category,
         total_revenue,
         profit,
         DENSE_RANK() OVER (ORDER BY profit DESC) as profit_rank
    FROM commerce
)
SELECT
   order_id,
   product_category,
   total_revenue,
   profit,
FROM commerce
WHERE profit_rank <= 5;
```
