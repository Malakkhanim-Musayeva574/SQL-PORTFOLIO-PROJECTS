```python
## Which product categories are causing a loss?

Select
     Category,
     Sub_Category,
     Round(Sum(Profit),2) AS Total_Profit
From superstore
Group BY Category, Sub_Category
Having SUM(Profit) < 0
ORDER BY Total_profit ASC;
```

```python
## Find customers who bought both Furniture and Technology in the same order?

SELECT DISTINCT
         a.Customer_ID,
         a.Customer_Name,
         a.Order_ID
From superstore a 
JOIN superstore b
   ON a.Order_ID = b.Order_ID
WHERE  a.Category = 'Furniture' AND b.Category = 'Technology';
```

```python
## How do cumulative (running total) sales grow month by month for each year?

SELECT
       EXTRACT(YEAR FROM Order_Date) AS sales_year,
       EXTRACT(MONTH FROM Order_Date) AS sales_month,
       SUM(sales) AS Monthly_sale
    FROM superstore
    GROUP BY EXTRACT(YEAR FROM Ordet_Date), EXTRACT(MONTH FROM Ordet_Date)
)
SELECT
     sales_year,
     sales_month,
     ROUND(SUM(Monthly_sales, 2) AS Monthly_Sales
     ROUND(SUM(Monthly_sales) OVER (partition by sales_year ORDER BY sales_Month), 2) AS cumulative_sales
From MonthlySales
ORDER BY sales_year, sales_month;
```

```python
## Categorize orders into "High Profit", "Medium Profit", and "Loss" and count them.

SELECT
     CASE
         WHEN Profit > 100 THEN 'High_PROFIT'
         WHEN Profit BETWEEN 0 AND 100 THEN 'MEDIUM PROFIT'
         ELSE 'LOSS'
    END AS Profit_Category,
    COUNT(Row_Id) AS Order_Count,
    ROUND(SUM(Sales),2) AS Total_Sales
From superstore
GROUP BY 
     CASE
         WHEN Profit > 100 THEN 'High Profit'
         WHEN Profit BETWEEN 0 AND 100 THEN 'Medium_Profit'
         ELSE 'Loss'
    END;
```

```python
## Categorize orders into "High Profit", "Medium Profit", and "Loss" and count them.

SELECT
   Profit_Category,
   COUNT(*) AS Order_Count,
   ROUND(SUM(Sales),2) AS Total_Sales
FROM(
     SELECT
      *,
       CASE
           WHEN Profit > 100 THEN 'High Profit'
           WHEN Profit BETWEEN 0 AND 100 THEN 'MEDIUM PROFIT'
           ELSE 'LOSS'
      END AS PROFIT_CATEGORY
    FROM superstore
) AS sub
GROUP BY Profit_Category;
```

```python
## Find products that were ordered together in the same Order Id.

SELECT
      a.Order_Id,
      a.Poroduct_Name AS Product_A,
      b.Product_Name AS Product_B
From superstore a
INNER JOIN superstore b
   ON a.Order_Id = a.Order_Id
WHERE a.Product_Id < b.Product_Id
LIMIT 5;
   
```

```python
## Which customers have placed orders in two different ship modes(e.g, Standard Class AND First Class)?

SELECT DISTINCT
     a.Customer_Id,
     a.Customer_Name,
FROM superstore a
INNER JOIN superstore b
   ON a.Customer_Id = b.Customer_Id
WHERE a.Ship_Mode = 'First Class' AND b.Ship_Name = 'Standard Class';
```
